```
def group_and_barplot(df=df,feat_groupby='param_marka-pojazdu', feat_agg='price_value', agg_funcs=[np.mean,np.median,np.size], feat_sort='mean',top=50,figsize=(10,15),subplots=True):
  return (
  df
  .groupby(feat_groupby)[feat_agg]
  .agg(agg_funcs)
  .sort_values(by=feat_sort,ascending=False)
  .head(top)
).plot(kind='bar', figsize=figsize,subplots=subplots)
```

```
df['price_log'] = np.log(df['price_value'])
df.head(5)          
```

```
group_and_barplot(feat_groupby=('param_marka-pojazdu'),feat_agg='price_log',feat_sort='median', figsize=(20,15));
```
